```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from plotnine import *
```

```python
# Styles with seaborn
sns.set_context("paper")
```

```python
df = pd.read_csv("behindKs.csv", header=0)
```

```python
Ks = [1,2,3,4,5]
```

##### Amparo

```python
df_amparo = pd.DataFrame(df[["AMPARO1", "TimeAMPARO1"]])
df_amparo.rename({"AMPARO1": "Value","TimeAMPARO1": "Time" },axis=1, inplace=True)
df_amparo["K"]="1"
df_amparo = df_amparo.dropna()
#df_amparo = df_amparo.drop(0)
df_amparo

for i in range(1,len(Ks)):
    df_aux = pd.DataFrame(df[[f"AMPARO{Ks[i]}", f"TimeAMPARO{Ks[i]}"]])
    df_aux.rename({f"AMPARO{Ks[i]}": "Value",f"TimeAMPARO{Ks[i]}": "Time" }, axis=1, inplace=True)
    df_aux["K"]=f"{Ks[i]}"
    df_aux = df_aux.dropna()
    #df_aux = df_aux.drop(0)
    df_amparo = pd.concat([df_amparo,df_aux], axis=0, )
    
maxim=df_amparo["Value"].max()
```

```python
plot_amparo = (

    ggplot(df_amparo, aes('Time', 'Value', color='K')) + 
    
    geom_point() +
    
    geom_line(aes(group='K')) +
    
    
    
    ggtitle(r'$k_\max$-performance Amparo ($1$ min)') +
    
    xlab("Time (s)") + 
    
    scale_x_log10()+
    
    ylim(14000, 22500) +
       
    geom_hline(yintercept=maxim, alpha=1/5, linetype='dashed', color='red') +
    
    theme_538() 
)

plot_amparo
```

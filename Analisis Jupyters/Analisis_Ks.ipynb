```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Styles with seaborn
sns.set_context("paper")
```

```python
df = pd.read_csv("behindKs.csv", header=0)
```

```python
df
```

```python
Ks = [1,2,3,4,5]
```

```python
plt.figure(figsize=(10,5))
plt.rc('text', usetex=True)
for K in Ks:
    plt.plot(df["TimeAMPARO"+f"{K}"], df["AMPARO"+f"{K}"], marker='o', ms=4, label = fr"$\alpha = ${K}")
    
        

plt.legend()
plt.grid(alpha=0.5)
plt.title(r"$k$-performance AMPARO (1 min)")

plt.show()
```
